```python
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
#Clean and complete dataset. Shape = (7895343, 52)
#df = pd.read_csv("/content/drive/My Drive/DS4AProject/CleanDB_FULL.csv")
df = pd.read_csv("CleanDB_FULL.csv")
```

#Adding the variable (Column) to know if the original dataset is born, fetal or death.
#born = 1, fetal = 2, defun = 3

```python
print(df.shape)
df["ORIGINAL_FILE"].unique()
```

```python
#Adding the column

df.loc[df['ORIGINAL_FILE'].str.find("nac")>-1, 'DS_ORIG'] = 1.0  
df.loc[df['ORIGINAL_FILE'].str.find("fetal")>-1, 'DS_ORIG'] = 2.0  
df.loc[df['ORIGINAL_FILE'].str.find("defun")>-1, 'DS_ORIG'] = 3.0  

print(df["DS_ORIG"].unique())
```

```python
print(df.shape)
df.tail()
```

```python
#Saving this Dataframe to not make this procedure again 
#dfDsOrigin.to_csv("CleanDB_FULL_DsOrigin.csv")
```

```python
print("Porcentaje de vivos en el df",(df[df["DS_ORIG"] == 1].shape[0]/df.shape[0])*100)
print("Porcentaje de fetales en el df",(df[df["DS_ORIG"] == 2].shape[0]/df.shape[0])*100)
print("Porcentaje de defun en el df",(df[df["DS_ORIG"] == 3].shape[0]/df.shape[0])*100)
```

#Dividing the data frame with a random sample to work with 20%

This is not for train and test. This division is done later, it is simply not to work with the 7 million data. It is done randomly and with a seed set to 40 for later repeatability.

```python
train_df, test_df = train_test_split(df, train_size=0.2, shuffle = True, random_state=40)
```

```python
print(train_df.shape)
train_df.tail()
```

Finding the percentage of each category of original dataset to ensure that we have enough representation of born, fetal and deceased

```python
print("Porcentaje de vivos en el df",(train_df[train_df["DS_ORIG"] == 1].shape[0]/train_df.shape[0])*100)
print("Porcentaje de fetales en el df",(train_df[train_df["DS_ORIG"] == 2].shape[0]/train_df.shape[0])*100)
print("Porcentaje de defun en el df",(train_df[train_df["DS_ORIG"] == 3].shape[0]/train_df.shape[0])*100)
```
